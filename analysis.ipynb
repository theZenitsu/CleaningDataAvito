```python
# Function to check for rental-related keywords in the title
def impute_type_de_vente(title, current_type):
    if pd.isnull(current_type):
        if any(keyword in str(title).lower() for keyword in ["louer", "loué", "location"]):
            return "louer"
    return current_type

# Apply function to 'type_de_vente' based on 'title'
df['type_de_vente'] = df.apply(lambda row: impute_type_de_vente(row['title'], row['type_de_vente']), axis=1)

# Now, fill remaining NaN values with 'louer' as a general fallback
df['type_de_vente'].fillna('louer', inplace=True)

# Verify if any NaN values remain in 'type_de_vente'
type_de_vente_missing_final = df['type_de_vente'].isnull().sum()

type_de_vente_missing_final
```
